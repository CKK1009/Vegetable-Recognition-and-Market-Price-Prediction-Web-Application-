```python
# all type of round cabbage named as round cabbage 
# all type of red onion named as red onion 
L = ['holland sweet potato', 'yellow / holland onion','red onion','green chili','red chili kulai / kulai hybrid',
    'garlic','coriander','okra','green bean / string bean','long beans', 'luffa','round cabbage','choy sum', 'long eggplant', 
     'cucumber','tomato', 'scallions']

pat = '|'.join(r"\b{}\b".format(x) for x in L)

price['new_name'] = price['vege_name'].str.extract('('+ pat + ')', expand=False, flags=re.I)
```

```python
price
```

```python
# summarize the price of vege of all states
vg_price = price.groupby('new_name').agg(unit=pd.NamedAgg(column="unit",aggfunc="unique"),
                                          avg_max=pd.NamedAgg(column="max_price", aggfunc="mean"),
                                          avg_min=pd.NamedAgg(column="min_price", aggfunc="mean"))
```

```python
vg_price
```

```python
# remove bracket
import re
vg_price['unit'] = vg_price['unit'].map(lambda x: re.sub(r'\W+', '', str(x)))

# reset index
vg_price = vg_price.reset_index(level=0)

# convert all string into lower case
vg_price['new_name']=vg_price['new_name'].str.lower()
vg_price['unit']=vg_price['unit'].str.lower()
```

```python
vg_price
```

```python
vg_price.columns = vg_price.columns.str.replace('new_name', 'vege_name')
```

```python
# function to get price range
def get_price_range(vege):
    min_price = round(price.avg_min[price.vege_name == vege].iloc[0],2)
    max_price= round(price.avg_max[price.vege_name == vege].iloc[0],2)
    vg_unit= price.unit[price.vege_name == vege].iloc[0]
    price_range = "RM " + str(min_price) + " to RM " + str(max_price) + " per " + vg_unit
    return price_range
    
```
